## Return Multiple Values
"return" values from the function. in java we can return only one value from the function but in python we can return multiple values

```python
# example for returning multiple values
def summul(a,b):# formal args
    c= a+b
    d= a*b
    return c,d

#Main Program
print(summul(10,20))
```

## Functions are first class objects

in python, functions are considered as objects. infact when we define a function, python will create an object. 
so we can pass function to another function as we pass object. and also we can return a function from another function

```python
def display(str):
    return 'Hai '+str

print(type(display))
x= display("Rama")
print(x)
```

```python
# you can define function inside another function. this is called nested function

def display(str):
    # Inside Function
    def message():
        return "How are you ? "
    result =  str + ',' + message()
    return result
    #display function ends

# main program
print(display("Krishna"))
```

## Pass by Value and References

```python
# pass object by reference
#in java and c we can pass variables to a function either Pass by Value or Pass by reference.
#But in python, everything is objects like integer, string, float..they are immutable objects but list and dictionaries are
#mutable so object  can be modified.

# Integer Class is IMMUTABLE
# when we try to modify the integer object inside the function, new object got created..samething will happen to float
# string and tuple
          
def modify(x):
    x=11
    print("Inside x=",x,"ID=",id(x))

#Main Program Starts
x= 10
print("Outside Before x=",x,"ID=",id(x))

# calling function
modify(x)

print("Outside After x=",x,"ID=",id(x))
print()
```

```python
# Since list is mutable object. passed object can be modified.
def modifyList(myList2):
    print("Inside myList=",myList2,"ID=",id(myList2))
    myList2[1]=99
    print("After update Inside myList=",myList2,"ID=",id(myList2))

myList=[0,1,2]
print("Outside Before myList=",myList,"ID=",id(myList))

# calling function
modifyList(myList)

print("Outside After myList=",myList,"ID=",id(myList))
print()
```

## Function Arguments

```python
#Formal and Actual arguments
# when we define function, we mention some parameter to receive data from outside. these are called FORMAL arguments
# when we CALL function, we pass some parameters to the function and these are called ACTUAL arguments

def sum(a,b): # a and b are formal arguments
    c=a+b
    print(c)
    
x,y=10,15
sum(x,y) # x and y are actual arguments

```

Actual arguments are of 4 types
1. positional --attach('New','york') - order is important
2. keyword -- grocery (item='sugar',price=50.75)--use the formal arguments name, order is not important
3. default arguments -- def grocery (item,,price=50.75)--when we define use some default value,if actual arguments is not there then default value will be used
4. variable length arguments--def add(farg,*args)---it can accept any number of arguments

```python
# for Positional Arguments

def attach(fword,sword):
    return fword+sword

#Main
attach("New",'York')
```

```python
#example for keyword arguments

def grocery(itemParam,priceParam):
    print("Item =",itemParam,"Price=",priceParam)    
#Main Program

item='Sugar'
price=100.45
grocery(priceParam=price,itemParam=item)
```

```python
#example for Default arguments

def groceryDefault(itemParam='sugar',priceParam=150.35):
    print("Item =",itemParam,"Price=",priceParam)  
    
item='Sugar'
groceryDefault('onion',100)
```

```python
# example for variable length arguments

def add(farg,*args):
    print(farg,args)
    sum=farg
    for i in args:
        sum+=i
    print(sum)
        
#main program
add(4,1,2)
```

## Local and Global Variables 
when we declare a variable inside a function it becomes local variable. scope is limited to that function only.
when we declare a variable outside a function it becomes GLOBAL variable. it can be accessed from entire program written below.
if the Same variable name given inside a function, then GLOBAL keyword can be prefixed to access global varible from inside a function
- You can pass group of elements to a function as parameters by passing a list
- function that calls itself is called recursive function.ex. factorial function.

```python
# example for local variable
def myfunction():
    aLocal=1
    aLocal+=1
    print(aLocal)
#main program starts
myfunction()
print(aLocal)#------------> cannot access outside. gives error
```

```python
# example for global variable
aGlobal=5

def myfunction2():
    bLocal=1
    bLocal+=1
    aGlobal=10 ## vakue got changed so it is different integer object
    print(aGlobal,bLocal)
    
#main program starts
myfunction2()
print(aGlobal)
```

## Anonymous Functions or Lamdas
A function without name is called Lambda functions.  functions are defined using Lambda (not using def)

lambda x: x*x

normal function returns values. but Lamda function returns "FUnction" so it should be assigned to function variable



```python
f= lambda x: x**2
print(f(5))
```

```python
# max of 2 numbers
max=lambda x,y :x if x>y else y
print(max(8,7))
```

```python
# use lambda function with filter functionx

# syntax for filter function is filter(function_name,sequence)--it applies all the elements to the function

#filter the even numbers

mylist=[2,3,4,5,6,7,8]

for i in mylist:
    if i%2==0:
        print(i)

print(list(filter(lambda x:(x%2==0),mylist)))

```

```python
# lamda with filter - example #2
# Take a list of numbers.  
my_list = [12, 65, 54, 39, 102, 339, 221, 50, 70, ] 
  
# use anonymous function to filter and comparing  
# if divisible by 13 or not 
result = list(filter(lambda x: (x % 13 == 0), my_list))  
  
# printing the result 
print(result) 
```

## Decorators

- decorator is a function that accepts the function and returns the function.decorators are useful to perform 
- additional process required by a function
- generators is a function that returns the sequence of values or elements. to return elements from a 

- generator we have to use yield statement.
- the value of the special variable '__name__' indicates whether the program is run as an individual program or as 
- as a module.if the value of the variable is '__main__' then the program is run directly as a program or 
- if it displays module name then it runs by a module.


```python
def first(msg):
    print(msg)    
first("Hello")

# Tag Function Name into another variable
second = first
second("Hello")
```

```python
# function as parameter
def inc(x):
    return x + 1

def dec(x):
    return x - 1

def operate(func, x):
    result = func(x)
    return result

#Main Program Starts
print(operate(inc,5))

print(operate(dec,5))
```

```python
# function can return another function

def is_called():
    def is_returned():
        print("Hello")
    return is_returned

# is_called function returns another function is_returned which is assigned to new
new = is_called()

#Outputs "Hello"
new()
```

```python
# Decorator
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner

def ordinary():
    print("I am ordinary")

fn_inner = make_pretty(ordinary)
fn_inner()
```

```python
# Decorator
def make_pretty(func):
    def inner():
        print("I got decorated")
        func()
    return inner

@make_pretty
def ordinary():
    print("I am ordinaryyy")

#fn_inner = make_pretty(ordinary)
ordinary()
```

```python
# Decorating function with Parameter

def smart_divide(func):
#number of porameters for inner function should be same as calling function
   def inner(a,b):
      print("I am going to divide",a,"and",b)
      if b == 0:
         print("Whoops! cannot divide")
         return

      return func(a,b)
   return inner

@smart_divide
def divide(a,b):
    return a/b

divide(8,0)

```

```python
def star(func):
    def inner(*args, **kwargs):
        print("*" * 30)
        func(*args, **kwargs)
        print("*" * 30)
    return inner

def percent(func):
    def inner(*args, **kwargs):
        print("%" * 30)
        func(*args, **kwargs)
        print("%" * 30)
    return inner

@star
@percent
def printer(msg):
    print(msg)
    
@star
@percent
def printerNew(msg1,msg2):
    print(msg1,msg2)
    
#Main Program Starts

printer("Hello")
printerNew("Hello","World")
```
