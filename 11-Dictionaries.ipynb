# Dictionaries in python

A dictionary represents a group of elements arranged in the form of key-value pair.
First element is considered as KEY and immediate next element as VALUE.
Key and Vaue is separated by :
All the key-Value pairs are inserted within 
ex:
mydict = { key1:value1,Key2-Value2 }


```python
# Create Dictionary with employee details and retrieve it 

mydict = {'EmpName':'Chandra','EmpId':200,'EmpSalary':9502.50}
print(mydict)

#Access value by dictionary key
print("Name of the Employee:",mydict['EmpName'])
print("ID of the Employee:",mydict['EmpId'])
print("Salary of the Employee:",mydict['EmpSalary'])
```

```python
#Operation on Dictionaries
mydict = {'EmpName':'Chandra','EmpId':200,'EmpSalary':9502.50}
# len function
print("Using Len function" , len(mydict))
```

```python
# Insert a new key value
mydict["Dept"]="IT"
print("After inserting the dept value", mydict)
```

```python
# modify the dictionary

mydict["EmpName"]="Sara"
print("After updated the empname value", mydict)
```

```python
# Deleting a key value pair
del mydict["EmpSalary"]
print("After the deleting the Salary value",mydict)
```

```python
#Clear method
mydict.clear()
print("After using clear()", mydict)
```

```python
# inserting the duplicate value
mydict["Dept"]="Finance"
print("After inserting the duplicate dept key=value", mydict)
```

```python
#copy method
mydict = {'EmpName':'Chandra','EmpId':200,'EmpSalary':9502.50}
#Alias
newdict = mydict

print("copied the dict in to newdict variable", mydict,"id",id(mydict))
print("copied the dict in to newdict variable", newdict,"id",id(newdict))

#Deep Copy
newdict = mydict.copy()

print("\ncopied the dict in to newdict variable", mydict,"id",id(mydict))
print("copied the dict in to newdict variable", newdict,"id",id(newdict))

```

```python
#fromkeys method - create new dictionary with keys from the other dict and with default values
keys = {'a', 'e', 'i', 'o', 'u' }
vowels = dict.fromkeys(keys,"vowel")
print(vowels)
```

```python
#get()
print(vowels.get("a"))
```

```python
# items()
print(vowels.items())
```

```python
### keys and values
print(vowels.keys())
print(vowels.values())
```

```python
# update- means append another dictionary with the exisitng one
vowels.update(mydict)
print(vowels)
```

```python
#pop()
vowels.pop("EmpName")
print("after using pop()",vowels)
```

```python
# converting lists to dictionary using zip()
countries=["USA","India","Germany","France"]
cities=["washington","New Delhi","Berlin","Paris"]
z=zip(countries,cities)
d=dict(z)
print(d)
```

```python
#ordered dictionary
from collections import OrderedDict
d=OrderedDict()
d[10]="A"
d[20]="B"
d[30]="C"
d[40]="D"
d[50]="E"
print(d)
```
