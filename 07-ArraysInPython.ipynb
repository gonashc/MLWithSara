## Array Creation

```python
# creating an integer array

import array as ar
a = ar.array('i',[4,6,2,9])
print(type(a))

for x in a:
    print(x)
```

```python
# creating array from another array
import array as ar
arr1=ar.array('d',[1.5,2.5,3.5])
print(arr1.typecode)
arr2=ar.array('d',(a*3 for a in arr1))

for i in arr2:
    print(i)
```

## Slicing Operations

```python
# slicing operation
import array as ar
x=ar.array('i',[10,20,30,40])
print(x[1:4])
print(x[:4])
print(x[1:])
print(x[-3:])
print(x[1:5:2])
```

## Array Methods

```python
# processing array
# Array class has several methods to process the Array
"""
.append
.extend
.fromfile
.tofile
.fromstring
.tostring
.fromlist
.tolist
.insert
.pop.
.remove
.reverse
"""

# A python program to understand various methods of array class
from array import *

arr= array('i',[10,20,30,40,50])
print('Original Array : ', arr)

#append 30 in the array arr:
arr.append(30)
arr.append(60)
print("After appending 30 and 60 are : ", arr)

# insert 99 at position 1
arr.insert(1,99)
print("After inserting 99 at the position 1 :", arr)

# remove 20 from the array
arr.remove(20)
print("After removing 20 from the array :",arr)

# remove the last element using pop method
arr.pop()
print("After removing the last element :", arr)

# finding the position of the element using index method
n=arr.index(30)
print("First occurence of element 30 is at ", n)

# convert array to list using .tolist() method
lst=arr.tolist()

print("list :",lst)
print("Array :", arr)
```

## Comparing Arrays

```python
# we can use the relational operators >,<=,==,!= to compare the arrays of same size. 
# it can compare the 2 arrays of same size and return the array of boolean values.
# use of any() and all()

import numpy as np
a = np.array([1,2,3,0])
b = np.array([0,2,3,1])
c = a>b
print(c)

# use of any() and all()
print("Check if ANY one element of array C is true---->", any(c))
print("Check if ALL one element of array C is true---->", all(c))

d = np.logical_and(a>0, a<4)
print("logical_and a>0 and a<4", d)
print(d)

d = np.logical_or(a>0, a<4)
print("logical_or a>0 and a<4", d) 
```

## Using WHERE 

```python
a = np.array([10,20,30,40,50],int)
b = np.array([1,21,3,40,51])

#if a is greater then take a else b
c= np.where(a>b, a,b)
print(c)
```

## Aliaing , Deep Copy and View

```python
#  Aliasing the array
b=a
print(a,b)
a[0]=99
print("***")
print(a,b)
b
```

```python
##Deep Copy
a = np.array([1,2,3,0])
b = a.copy()
print("Deep Copy")
print(a)
print(b)
print(b is a)
b[0]=99
print(b)
print(a)
```

```python
# creating view

b=a.view() # it is also called as shallow copy.it is different from alias. it has separate memory. 
print(type(a))
print(type(b))

print(b is a)
print(b.base is a)
```

## n-dimensions and attributes 

```python
# Single dimensional Array
a=np.array([1,2,3,4])
print(a)
b=np.array([1,
            2,
            3,
            4])
print(b)
print(b.ndim)
# 2 dimensional arrays 2x4

c=np.array([[1,2,3],[4,5,6]])
print(c)
print(c.ndim)
# 3 dimensional arrays is combination of several 2d arrays

d=np.array([[[1,2,3],[4,5,6]],[[3,2,1],[6,5,4]]])
print(d)
print(d.ndim)
```

```python
# Attributes of array

# ndim  represents number of dimensions 1 for 1d array, 2 for 2d arrays, 3 for 3d

# shape attribute - gives the shape of an array

# Single dimensional Array
a=np.array([1,2])
print(a.shape)
# 2 dimensional arrays 2x4

c=np.array([[1,2,3],[4,5,6]])
print(" 2d Array",c)
print(c.shape)
# 3 dimensional arrays is combination of several 2d arrays

d=np.array([[[1,2,3],[4,5,6]],[[3,2,1],[6,5,4]]])
print(d)
print(d.shape)

# Size Attribute gives number of elements in the array

print(d.size)

#itemsize attribute gives the memory size of the array element in bytes
print(d.itemsize)

#reshape()

#reshape method is useful to change the shape of the array
# the new array should have the same number of elements of the original array

arr1= np.arange(10)
print(arr1)
arr2=arr1.reshape(5,2)
print(arr2)

# Flatten method
print(arr2.flatten())

```

```python
# multidimensional arrays
# 2d and 3d arrays are called as multidimensional arrays
# 2d array is also called as Matrix. mxn m is number of rows and n is number of columns

# multidimensional array can be created using following methods

# 1. array()
# 2. ones() and zeros()
# 3. eye()
# 4. reshape()

a = np.array([[1,2,3],[4,5,6]])
print(a)
b = np.ones((2,3),int)
print(b)
c = np.zeros((2,3),int)
print(c)
```
